```python
from detecto import utils
from detecto import visualize
from detecto.core import Model
from detecto import utils,core
import os
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import random
from skimage import io
```

```python
path = 'mydata'
folders = [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]
annotations_path = os.path.join(path, 'annotations/train')
imagess = os.path.join(path, 'images/train')
test_annotations = os.path.join(path, 'annotations/test')
test_images = os.path.join(path, 'images/test')
utils.xml_to_csv(annotations_path, 'train.csv')
utils.xml_to_csv(test_annotations, 'val.csv')
dataset = core.Dataset('train.csv',imagess)
val_dataset = core.Dataset('val.csv', test_images)
loader = core.DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
# Do not run this cell for testing. The model is already trained and saved.
model = Model(['finger'])
losses = model.fit(loader, val_dataset, epochs=20, verbose=True)
```

```python
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
model.save('finger_detector.pth')
```

```python
def calculate_iou(boxes1, boxes2):
    x1 = np.maximum(boxes1[:, 0], boxes2[:, 0])
    y1 = np.maximum(boxes1[:, 1], boxes2[:, 1])
    x2 = np.minimum(boxes1[:, 2], boxes2[:, 2])
    y2 = np.minimum(boxes1[:, 3], boxes2[:, 3])
    intersection = np.maximum(x2 - x1, 0) * np.maximum(y2 - y1, 0)
    area1 = (boxes1[:, 2] - boxes1[:, 0]) * (boxes1[:, 3] - boxes1[:, 1])
    area2 = (boxes2[:, 2] - boxes2[:, 0]) * (boxes2[:, 3] - boxes2[:, 1])
    union = area1 + area2 - intersection
    iou = intersection / union
    return iou
```

```python
def evaluate(model, dataset,iou_threshold):
    total_true_positives = 0
    total_false_positives = 0
    total_false_negatives = 0
    total_iou = 0
    num_boxes = 0
    f1_scores = []
    avg_ious = []
    
    for index in range(len(dataset)):
        image, target = dataset[index]
        true_labels = np.array(target['labels'])
        true_boxes = np.array(target['boxes'])
        
        labels, boxes, scores = model.predict(image)
        selected_boxes = boxes[scores > 0.2]
        selected_boxes = np.array(selected_boxes)
        if selected_boxes.shape != true_boxes.shape:
            continue
        
        iou = calculate_iou(selected_boxes, true_boxes)
        true_positives = np.sum(iou >=iou_threshold, axis=0)
        false_positives = np.sum(iou < iou_threshold, axis=0)
        false_negatives = np.sum(np.sum(iou > iou_threshold, axis=0) == 0)
        
        total_true_positives += np.sum(true_positives)
        total_false_positives += np.sum(false_positives)
        total_false_negatives += false_negatives
        total_iou += np.sum(iou)
        num_boxes += iou.shape[0]
    
        precision = total_true_positives / (total_true_positives + total_false_positives)
        recall = total_true_positives / (total_true_positives + total_false_negatives)
        f1_score = 2 * precision * recall / (precision + recall)
        avg_iou = total_iou / num_boxes
        f1_scores.append(f1_score)
        avg_ious.append(avg_iou)
    return f1_scores, avg_ious
```

```python
scores, ious = evaluate(model, val_dataset, iou_threshold=0.6)
mean_f1_score = np.mean(scores)
mean_avg_iou = np.mean(ious)
print(f'F1 score: {mean_f1_score:.4f},  IOU :{mean_avg_iou:.4f}')
```
